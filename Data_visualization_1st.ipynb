```
data = pd.read_csv(io.BytesIO(uploaded['company_sales_data.csv']))
```

```
display(data.info(), data.head())
```

```
total_profit = data['total_profit']
month = data['month_number']
plt.plot(month, total_profit)
```

```
c1 = data.iloc[:,1]
c2 = data.iloc[:,2]
c3 = data.iloc[:,3]
c4 = data.iloc[:,4]
c5 = data.iloc[:,5]
c6 = data.iloc[:,6]
```

```
plt.plot(month, c1, color='violet', label='facecream')
plt.plot(month, c2, color='indigo', label='facewash')
plt.plot(month, c3, color='blue', label='toothpaste')
plt.plot(month, c4, color='green', label='bathingsoap')
plt.plot(month, c5, color='red', label='shampoo')
plt.plot(month, c6, color='orange', label='moisturizer')
plt.legend()
plt.show()
```

```
product = ['facecream','facewash', 'toothpaste', 'bathingsoap','shampoo','moisturizer']
total = [sum(c1),sum(c2),sum(c3),sum(c4),sum(c5),sum(c6)]
plt.pie(total, labels = product)
plt.show()
```
